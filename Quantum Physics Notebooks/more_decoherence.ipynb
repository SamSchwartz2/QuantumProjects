```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

## Problem 1

### 1b)

```python
#params/definitions
delta = 0
omega = np.pi*2
gamma = .1*omega
psi0 = basis(2,0) #inital state is |0>
H = -.5 * delta * sigmaz() + .5* omega * sigmax()
times = np.linspace(0,15,1000)

emiss_op = np.sqrt(gamma)*destroy(2) #define the photon emission operator (collapse operator)
excited_pop = basis(2,0)*basis(2,0).dag() #we want to calculate the excited state population (expectation operator)

result = mesolve(H,psi0,times,emiss_op,excited_pop)
excite = result.expect[0]


plt.plot(times,excite, label = 'Excited state population')
plt.xlabel('Time (s)')
plt.ylabel('Population of |0>')
plt.ylim(0,1)
plt.grid(True)
plt.title('Population of Excited state over time (Γ=.1Ω)')
```

### 1d)

```python
data = np.loadtxt("Ba133_SRdata.txt")
delta_vals = []
R_vals = []
for i in range(len(data)):
    delta_vals.append(data[i][0])
    R_vals.append(data[i][1])
    
def scatter_rate(delta, gamma, s): #defines our photon scattering rate function
    top = 2*gamma*delta*s
    bottom = gamma + 2*delta*s + ((2*delta)/gamma)**2
    return top/bottom

gamma_guess = 100 
s_guess = 1
init_guess = [gamma_guess, s_guess] #this is our first/base case for curve fitting

fit = np.polyfit(delta_vals, R_vals, deg=7) # a 7 degree poly fit seems to be a solid midpoint between overfitting and accurate oscillation representation
p=np.poly1d(fit) #p is the function that maps delta vals to R vals


plt.figure(figsize=(10, 6))
plt.scatter(delta_vals, R_vals, label='Experimental Data')
plt.plot(delta_vals, p(delta_vals), label='Fit: $R(\\Delta)$', color='red')
plt.xlabel('Detuning (Δ)', fontsize=14)
plt.ylabel('Photon Scattering Rate (R)', fontsize=14)
plt.title('Photon Scattering Rate vs Detuning', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```
