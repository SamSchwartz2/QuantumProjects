```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
```

## Problem 4: Decoherence in Qutip

### 4a)

```python
delta = 0
omega = 2*np.pi

H = (-1*delta/2)*sigmaz() + (omega/2)*sigmax()
psi0 = basis(2,1) #ground state = |1>

times = np.linspace(0,5,250)

result = mesolve(H, psi0, times, [], [])
excited_proj = basis(2,0) * basis(2,0).dag()
excited_pop = expect(excited_proj, result.states)

plt.plot(times, excited_pop)
plt.xlabel("Time (s)")
plt.ylabel("Probability of Being in Excited State |0>")
plt.title("4a) Excited state population vs. time")
```
