## Problem 2: Displacement operator

## 2a:

### i) alpha = 1

```python
ket0 = basis(2,0)
alpha = 1
disp_0 = displace(2,alpha) * ket0
print(disp_0)

x_op = position(2) #define x as the position operator
p_op = momentum(2) #define p as the momentum operator

exp_x1 = expect(x_op, disp_0) #calculate <x>
exp_p1 = expect(p_op, disp_0) #calculate <p>
print(exp_x1, exp_p1)

plt.figure(figsize=(8, 8))
plt.quiver(0, 0, exp_x1, exp_p1, angles='xy', scale_units='xy', scale=1, color='black') #plot the vector from (0,0) to the point in x,p space
plt.scatter(exp_x1, exp_p1, label='Phase Space Trajectory of <x>, <p>)', color='black') #plot the point itself in x,p space
plt.title("Phase Space Trajectory of <x>, <p>")
plt.xlabel('<x>')
plt.ylabel('<p>')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.grid(True)
plt.show()

```

## 2b:

### i) alpha = 1

```python
m = 1
w = 2*np.pi
hbar = 1
times = np.linspace(0,1,100)

a = destroy(2)
adag = create(2)
x_op = position(2)
p_op = momentum(2)
H0 = hbar * w * (adag*a + .5) #define H0

def time_ev_op(t): #define U(t)
    return propagator(H0, t, unitary_mode='dense')

alpha = 1
psi_0 = displace(2, alpha) #define our initial state, the coherent state w/ alpha = 1

exp_x = []
exp_p = []

for t in times: #get the time evolution of psi_0
    Ut = time_ev_op(t)
    psi_t = Ut * psi_0
    exp_x.append(expect(x_op, psi_t))
    exp_p.append(expect(p_op, psi_t))


plt.figure(figsize=(8, 8))
plt.plot(exp_x, exp_p)
plt.xlabel('x')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.ylabel('p')
plt.title('Expectation Values of Position and Momentum Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

# Problem 3: Ramsey Spectroscopy w/ QHO

### a) P(|0>) as a function of tw

```python
m = 1
w = np.pi * 2
hbar = 1
alpha1 = 1
alpha2 = -1
vaccum = basis(2,0) #|0> is the vaccum state
a = destroy(2)
adag = create(2)
times = np.linspace(0,3,300)

H0 = hbar * w * (adag*a + .5) #define the Hamiltonian
disp_alpha1 = displace(2,alpha1) #displacement operator calculation for alpha1
disp_alpha2 = displace(2,alpha2) #displacement operator calculation for alpha2
psi0 = disp_alpha1 * vaccum #initial state

def time_ev_op(t): #define U(t)
    return propagator(H0, t, unitary_mode='dense')


p0_arr = []
for t in times:
    Ut = time_ev_op(t)
    psi1 = Ut * psi0 #This is the first displacement 
    psi2 = disp_alpha2 * psi1 #The second displacement, occuring when t = tw
    temp_prob = abs(psi2[0])**2
    p0_arr.append(temp_prob)

plt.figure(figsize=(10, 6))
plt.plot(times, p0_arr, label='Probability of being in the vacuum state')
plt.xlabel('Time')
plt.ylabel('Probability of |0>')
plt.ylim(0,1)
plt.title('Probability of Being in the Vacuum State as a Function of time')
plt.grid(True)
plt.legend()
plt.show()
```

### b) Set tw = .5

```python
m = 1
w = np.pi * 2
hbar = 1
alpha1 = 1
alpha2 = -1
vaccum = basis(2,0) #|0> is the vaccum state
a = destroy(2)
adag = create(2)
times = np.linspace(0,3,300)

H0 = hbar * w * (adag*a + .5) #define the Hamiltonian
disp_alpha1 = displace(2,alpha1) #displacement operator calculation for alpha1
disp_alpha2 = displace(2,alpha2) #displacement operator calculation for alpha2
psi0 = disp_alpha1 * vaccum #initial state

def time_ev_op(t): #define U(t)
    return propagator(H0, t, unitary_mode='dense')

x_op = position(2)
p_op = momentum(2)
exp_x = []
exp_p = []
tw = .5
for t in times:
    if t < tw:
        Ut = time_ev_op(t)
        psi_ev = Ut * psi0 #This is the first displacement
    else:
        psi2 = disp_alpha2 * psi_ev #The second displacement, occuring when t = tw
        Ut = time_ev_op(t-tw)
        psi_ev = Ut * psi2
    exp_x.append(expect(x_op,psi_ev))
    exp_p.append(expect(p_op,psi_ev))

plt.figure(figsize=(10, 5))
plt.plot(times, exp_x)
plt.xlabel('Time')
plt.ylabel('<x>')
plt.title('Time evolution of x with tw = .5 seconds')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(times, exp_p)
plt.xlabel('Time')
plt.ylabel('<p>')
plt.title('Time evolution of p with tw = .5 seconds')
plt.grid(True)
plt.legend()
plt.show()
```
